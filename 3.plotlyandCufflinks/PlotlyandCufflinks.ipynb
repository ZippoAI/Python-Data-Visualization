```python
import seaborn as sns

import pandas as pd

import numpy as np

import cufflinks as cf

from plotly.offline import iplot


import plotly.express as px


```

```python
tip = sns.load_dataset('tips')

tip
```

```python
tip['total_bill'].iplot()
```

```python
px.line(tip, y='total_bill')
```

```python

df = tip.groupby('day', as_index=False)['tip'].mean()
px.bar(df, x='day', y='tip')
```

## Scatter Plot

```python
px.scatter(tip, x = 'total_bill', y='tip')
```

## Box Plot

```python
df1 = tip[['total_bill', 'day']]

px.box(df1, x='total_bill', y = 'day')
```

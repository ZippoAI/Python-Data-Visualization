## Some basic Questions

### Which team won the most matches?

```python
match_wins = df['match_winner'].value_counts()

sns.barplot(y=match_wins.index, x=match_wins.values, palette='rainbow')
plt.title('Most Match Win By Team')
plt.show
```

### Toss Decision Trends

```python
counts = df['toss_decision'].value_counts()

print(f"Total {counts.sum()} choose toss decision")
print(f"total {counts.get('Bat', 0)} choose bat")
print(f"total {counts.get('Field', 0)} choose bat")

sns.countplot(x=df['toss_decision'], palette="mako", hue=df['toss_decision'])
plt.title('Toss Decision Trend')
```

## Key player performance

1. Most "Player of the match" Awards

```python
df['player_of_the_match'].value_counts().head()
```

```python
award_count = df['player_of_the_match'].value_counts().head(15)
sns.barplot(y=award_count.index, x=award_count.values, palette='mako', hue=award_count.index)
plt.title('Top 15 Players of the match')
plt.xlabel('Number of awards')
plt.ylabel('Name of player')
```

```python
sns.countplot(
    data=df,
    y='player_of_the_match',
    order=df['player_of_the_match'].value_counts().head(15).index
    
)
plt.title('Top 15 Players of the match')
plt.xlabel('Number of awards')
plt.ylabel('Name of player')
```

```python
award_count = df['player_of_the_match'].value_counts()
total = award_count.sum()

top_n = 15
award_top = award_count.head(top_n)
percentages = (award_top / total) * 100
```

```python

plt.figure(figsize=(8, 6))

ax = sns.barplot(
    x=award_top.values,
    y=award_top.index
)

plt.title("Top 15 Players of the Match")
plt.xlabel("Number of Awards")
plt.ylabel("Player")

for i, (count, pct) in enumerate(zip(award_top.values, percentages)):
    ax.text(
        count + 0.05,
        i,
        f"{pct:.1f}%",
        va='center'
    )
```

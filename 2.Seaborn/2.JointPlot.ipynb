# Joint Plot (Seaborn)

## What is a Joint Plot?
A **joint plot** is used to visualize the relationship between **two quantitative variables**  
along with their **individual distributions** in a single figure.

It combines:
- A relationship plot (center)
- Distribution of X variable (top)
- Distribution of Y variable (right)

So instead of 3 plots, you get **everything in one view**.


## Why do we use a Joint Plot?
We use a joint plot to:
- Understand how two variables are related
- Observe trends (positive / negative relationship)
- Detect outliers
- See how each variable is distributed

This is very useful during **Exploratory Data Analysis (EDA)** in ML.


## When should you use it?
Use a joint plot when:
- Both variables are **quantitative**
- You want to check correlation or dependency

Examples:
- `total_bill` vs `tip`
- `height` vs `weight`


## Basic Syntax


```python
sns.jointplot(data = df, x = 'total_bill', y='tip' , kind='scatter')

```

```python
sns.jointplot(data = df, x = 'total_bill', y='tip' , kind='hex')
```

```python
sns.jointplot(data = df, x = 'total_bill', y='tip' , kind='reg')
```

```python
sns.jointplot(data = df, x = 'total_bill', y='tip' , kind='kde')
```

## How to save this

```python
dj = sns.jointplot(data = df, x = 'total_bill', y='tip' , kind='scatter')
dj.figure.savefig('no1.png', dpi=300, bbox_inches='tight') # WE can increase the quality by increasing the dpi

# we can do transparent background too
dj.figure.savefig('no2.png', dpi=300, bbox_inches='tight', transparent=True)
```

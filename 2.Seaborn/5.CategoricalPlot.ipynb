# Categorical Plot

## What is a Categorical Plot?
A **categorical plot** is used to visualize **categorical data** (non-numeric labels) against
- numerical values **or**
- counts

Example of categorical data:
- **gender**
- **day**
- **smoker**
- **type**

---

## Why do we use a Categorical Plot?
We use categorical plots to:
- **Compare categories**
- See **differences between groups**
- Analyze **patterns across categories**

In AI/ML:
Categorical plots help understand **how categories affect numerical values**.

---

## Types of Categorical Plots in Seaborn
- **barplot**
- **countplot**
- **boxplot**
- **violinplot**
- **stripplot**
- **swarmplot**
- **pointplot**

---



# Countplot

```python
sns.countplot(df['sex'])
```

```python
# if you want to make it x axis

sns.countplot(x = df['sex'])
plt.show()
```

## Lets check how many males are smoking and how many females are smoking

```python
sns.countplot(x = df['sex'], hue=df['smoker'])
```

# Bar plot

```python
sns.barplot(x = df['sex'], y=df['total_bill'])
```

```python
sns.barplot(x = df['sex'], y=df['total_bill'], estimator=np.sum)
plt.title('Total bill')
```

```python
sns.barplot(x=df['sex'], y=df['tip'], estimator=np.sum)
plt.title('Tips')
```

# Boxplot

```python
sns.boxplot(x=df['tip'], y=df['day'], palette='rainbow')
plt.show()
```

## Boxplot and Violinplot

Boxplot and violinplot help us understand how data is spread for different categories.

**Boxplot**

Shows minimum, maximum, median, and quartiles

Helps compare values between groups

Outliers are shown as separate points

**In short:**
ðŸ‘‰ Boxplot shows summary of data using boxes and lines.

---

**Violinplot**

Shows the shape of the data distribution

Wider parts mean more data values there

Combines boxplot + distribution view

**In short:**
ðŸ‘‰ Violinplot shows how data is spread and where values are concentrated.

---

**When to use**

Use **boxplot** when you want quick comparison

Use **violinplot** when you want to see data distribution clearly

```python
sns.violinplot(x = df['day'], y=df['total_bill'])
plt.show()
```

## Stripplot and Swarmplot (Easy)

**Stripplot**

Shows each data point as a dot

One axis is category, the other is value

Points can overlap

**In short:**
ðŸ‘‰ Stripplot shows all data points, but they may overlap.

---

**Swarmplot**

Same as stripplot, but points do not overlap

Dots are adjusted sideways for clarity

Shows data distribution more clearly

**In short:**
ðŸ‘‰ Swarmplot spreads points so we can see them clearly.

--- 

**Difference**:

**Stripplot**â†’ simple, fast, overlaps allowed

**Swarmplot** â†’ clearer, no overlap, slower for large data

### Strip Plot

```python
sns.stripplot(x = 'day', y='total_bill', data=df)
plt.show()
```

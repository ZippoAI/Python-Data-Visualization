# Rug Plot

## What is a Rug Plot?
A **rug plot** shows **each individual data point** as a **small line (tick)** on an axis.

Each tick represents **one value** in the dataset.

---

## Why do we use a Rug Plot?
We use a rug plot to:
- **See exact data points**
- **Understand data density**
- **Find clusters**
- **Detect outliers**

Rug plots are usually used **along with other plots**, not alone.

---

## How to Read a Rug Plot
- **Many ticks close together** → High data density  
- **Few ticks far apart** → Low data density  
- **Single ticks alone** → Possible outliers  

---

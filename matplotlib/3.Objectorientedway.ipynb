```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)
plt.title('Title')
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
```

# Object Oriented Way

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x**2, x**2)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])

# or 

'''
fig.add_axes([
    0.1,  # left    10% from the left edge of the figure
    0.1,  # bottom  10% from the bottom edge of the figure
    0.8,  # width   axis takes 80% of the figure's width
    0.8   # height  axis takes 80% of the figure's height
])
'''

axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)

axis2 = fig.add_axes([0.3,0.4,0.3,0.3])
axis2.plot(x,y)
```

# Object Oriented Way


Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(12,8), dpi=100)
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])

# or 

'''
fig.add_axes([
    0.1,  # left    10% from the left edge of the figure
    0.1,  # bottom  10% from the bottom edge of the figure
    0.8,  # width   axis takes 80% of the figure's width
    0.8   # height  axis takes 80% of the figure's height
])
'''

axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)

axis2 = fig.add_axes([0.3,0.4,0.3,0.3])
axis2.plot(x,y)
```

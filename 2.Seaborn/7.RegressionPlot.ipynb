```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

## lmplot()

```python
sns.lmplot(data=tips, x='total_bill', y='tip')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex', palette='coolwarm')
```

## We can change the markers too

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex', markers=['o', 'v'])
```

## We can change the size

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex', markers=['o', 'v'], scatter_kws={'s':200}) 
# s mean size so S:100 means size = 100
```

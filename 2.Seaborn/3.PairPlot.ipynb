# üìä Pair Plot

## What is a Pair Plot?

A pair plot is a grid of graphs that shows relationships between multiple numerical columns in a dataset all at once.

Think of it as:

‚ÄúShow me how every number relates to every other number.‚Äù

## What does it contain?

A pair plot automatically creates:

Scatter plots ‚Üí for every pair of columns

Histograms (or KDE plots) ‚Üí on the diagonal (each column‚Äôs own distribution)

**So you get**:

Column vs Column

For all combinations

In one single view

## Why do we use Pair Plot?

We use a pair plot to quickly understand data patterns.

**It helps to:**

See relationships between variables

Spot trends (positive / negative relation)

Find clusters

Detect outliers

Get a quick overview before deeper analysis

In AI/ML:

Pair plots help you decide which features matter and how they behave together.

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

# Distribution Plot

We use **distribution plots** to visualize the distribution of **quantitative data**.

The following plots are commonly used to visualize the distribution of a dataset:

- **distplot / histplot**
- **jointplot**
- **pairplot**
- **rugplot**


```python

df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

### Histogram

```python

sns.displot(df['total_bill'])
```

```python
# We can also mention the bin

sns.displot(df['total_bill'], bins = 5)
```

### Kernel density in Histogram

```python
sns.histplot(df['total_bill'],kde = True)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],kde = True)
```

```python
sns.displot(data=df, x="total_bill", hue="sex", col="size", kind="hist")

# total_bill	tip	sex	smoker	day	time	size
```

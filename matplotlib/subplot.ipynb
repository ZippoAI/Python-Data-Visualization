```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)
plt.title('Title')
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x**2, x**2)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y)
```
